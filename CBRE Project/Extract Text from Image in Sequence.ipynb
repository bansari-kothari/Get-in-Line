```python
import cv2
import re 
from pytesseract import pytesseract, Output
#Define path to tessaract.exe
path_to_tesseract = r'C:/Program Files/Tesseract-OCR/tesseract.exe'
#Point tessaract_cmd to tessaract.exe
pytesseract.tesseract_cmd = path_to_tesseract
```

```python
# get grayscale image
def get_grayscale(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# noise removal
def remove_noise(image):
    return cv2.medianBlur(image, 1)
 
#thresholding
def thresholding(image):
    return cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
```

```python
custom_config = r'--oem 3 --psm 11'

def processImage(img):
    details = pytesseract.image_to_data(img, output_type=Output.DICT, config=custom_config, lang='eng')
    text = pytesseract.image_to_string(img, config=custom_config).splitlines()
    final = []
    s = ''
    for l in text:
            if l.endswith('.') or l.endswith('!') or l.endswith('?'):
                final.append(s + ' ' + l)
                s = ''
            else:
                s = s+ ' '+ l
    return final
```

```python
extracted_text = []
for i in range(1,22):
    path_to_image = 'comic_image/cartoon'+str(i)+'.png'
    img = cv2.imread(path_to_image)
    gray = get_grayscale(img)
    threshold_img = remove_noise(gray)
    extracted_text.append(processImage(threshold_img))
extracted_text
```

```python
import numpy as np
for i in extracted_text:
    for j in i:
        print(j, end=" \n")
    print()
```
